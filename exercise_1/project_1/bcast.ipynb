# Broadcast Benchmark

# Plotting Latency by Processes for each algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('bcast_thin_results/thin_bcast_results.csv')

# Plotting the data
fig, ax = plt.subplots(figsize=(10, 6))

# List of algorithms to plot
algorithms = ['basic_linear', 'chain', 'pipeline']

for algo in algorithms:
    # Filtering data for each algorithm
    filtered_data = data[data['Algorithm'] == algo]
    # Plotting each algorithm's latency vs. processes
    ax.plot(filtered_data['Processes'].unique(), filtered_data.groupby('Processes')['Latency'].mean(), label=algo)

ax.set_title('Latency by Number of Processes for Each Algorithm')
ax.set_xlabel('Number of Processes')
ax.set_ylabel('Average Latency (ms)')
ax.legend()
plt.axvline(x=24, color='red', linestyle='--')
plt.axvline(x=12, color='gray', linestyle='--')
plt.axvline(x=36, color='gray', linestyle='--')
#plt.grid(True)
plt.show()
```

# Latency by processes for each message size for `basic_linear` algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('bcast_thin_results/thin_bcast_results.csv')

# Filter data for the "basic_linear" algorithm
filtered_data = data[data['Algorithm'] == 'basic_linear']

# Plotting the data
fig, ax = plt.subplots(figsize=(12, 8))

# Get unique sizes and sort them
sizes = sorted(filtered_data['Size'].unique())

# Generate a color map
cm = plt.get_cmap('rainbow')
ax.set_prop_cycle('color', [cm(1. * i / len(sizes)) for i in range(len(sizes))])

for size in sizes:
    # Filter data for each size
    size_data = filtered_data[filtered_data['Size'] == size]
    # Plotting latency vs. processes for each size
    ax.plot(size_data['Processes'], size_data['Latency'], label=f'Size {size} bytes', marker='o')

ax.set_title('Latency by Number of Processes for Each Message Size (Basic Linear Algorithm)')
ax.set_xlabel('Number of Processes')
ax.set_ylabel('Latency (ms)')
ax.legend(title="Message Size", bbox_to_anchor=(1.05, 1), loc='upper left')
# plt.tight_layout()
plt.axvline(x=24, color='red', linestyle='--')
plt.axvline(x=12, color='gray', linestyle='--')
plt.axvline(x=36, color='gray', linestyle='--')
#plt.yscale('log')
plt.show()
```

# Latency by processes for each message size for `chain` algorithm

```python
# Filter data for the "chain" algorithm
filtered_data = data[data['Algorithm'] == 'chain']
basic_linear_data = data[data['Algorithm'] == 'basic_linear']

# Plotting the data
fig, ax = plt.subplots(figsize=(12, 8))

# Get unique sizes and sort them
sizes = sorted(filtered_data['Size'].unique())

# Generate a color map
cm = plt.get_cmap('rainbow')
ax.set_prop_cycle('color', [cm(1. * i / len(sizes)) for i in range(len(sizes))])

for size in sizes:
    # Filter data for each size
    size_data = filtered_data[filtered_data['Size'] == size]
    basic_size_data = basic_linear_data[basic_linear_data['Size'] == size]
    # Plotting latency vs. processes for each size
    ax.plot(size_data['Processes'], size_data['Latency'], label=f'Size {size} bytes', marker='o')
    ax.plot(basic_size_data['Processes'], basic_size_data['Latency'], marker='o', color='lightgray')

ax.set_title('Latency by Number of Processes for Each Message Size (Chain Algorithm)')
ax.set_xlabel('Number of Processes')
ax.set_ylabel('Latency (ms)')
ax.legend(title="Message Size", bbox_to_anchor=(1.05, 1), loc='upper left')
# plt.tight_layout()
plt.axvline(x=24, color='red', linestyle='--')
plt.axvline(x=12, color='gray', linestyle='--')
plt.axvline(x=36, color='gray', linestyle='--')
#plt.yscale('log')
plt.show()
```
